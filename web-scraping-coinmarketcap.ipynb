```python
volume_24h = []
price = []
market_cap = []
for x in coins:
    #print(x['symbol'], x['quote']['USD']['volume_24h'], x['quote']['USD']['market_cap'])
    volume_24h.append(x['quote']['USD']['volume_24h'])
    price.append(x['quote']['USD']['price'])
    market_cap.append(x['quote']['USD']['market_cap'])
```

```python
column_names = ["name", "symbol", "price","volume_24h", "market_cap"]
coins_df = coins_df.reindex(columns=column_names)
```

```python
for labels, content in coins_df.items():
    if pd.api.types.is_string_dtype(content):
        print(labels)
```

```python
for labels, content in coins_df.items():
    if pd.api.types.is_string_dtype(content):
        coins_df[labels] = content.astype("category")
```

```python
coins_df.info()
```

```python
for labels, content in coins_df.items():
    if not pd.api.types.is_numeric_dtype(content):
        coins_df[labels] = pd.Categorical(content).codes+1
```

```python
coins_df.info()
```

```python
x = coins_df.drop("price", axis = 1)
y = coins_df["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test  = train_test_split(x,y, test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.get_params()
```

```python
model.fit(x_train, y_train)
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
model.score(x_test, y_test)
```
